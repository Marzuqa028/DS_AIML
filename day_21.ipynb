Types of Machine Learning
1. Supervised Learning

Supervised Learning is a type of Machine Learning where the data contains input features along with correct output labels. The model learns by mapping inputs (X) to known outputs (y). The goal is to make accurate predictions on new, unseen data.
Data has answers/labels.
Examples:
Predicting house prices
Spam email detection

2. Unsupervised Learning

Unsupervised Learning is used when the dataset does not contain labeled outputs. The model tries to discover hidden patterns, structures, or clusters in the data.
Data has no labels.
Examples:
Customer segmentation
Market basket analysis

3. Reinforcement Learning

Reinforcement Learning is a type of learning where an agent interacts with an environment and learns by receiving rewards or penalties based on its actions. The goal is to maximize cumulative reward over time.
Learning happens via rewards and penalties.
Examples:
Self-driving cars
Game-playing AI
| Learning Type     | Data Type      | Goal                          | Real-World Examples                            |
| ----------------- | -------------- | ----------------------------- | ---------------------------------------------- |
| **Supervised**    | Labeled data   | Predict output from input     | Spam Detection, House Price Prediction         |
| **Unsupervised**  | Unlabeled data | Find hidden patterns/clusters | Customer Segmentation, Fraud Pattern Detection |
| **Reinforcement** | Reward-based   | Maximize long-term reward     | Self-Driving Cars, Robotics Navigation         |
