ðŸ“Š Customer Analytics Dataset Overview

The Customer Analytics dataset contains information about 255 customers, with each row representing a single customer. The dataset includes 14 features describing customer demographics, income details, purchasing behavior, and engagement patterns.

Key attributes include demographic information such as Age, Gender, City, Education, and Marital Status. It also contains behavioral and financial features such as Annual Income, Spending Score, Purchase Frequency, Online Visits Per Month, Returned Items, Preferred Device, and Last Purchase Amount.

This dataset can be used to analyze customer purchasing behavior, understand spending trends, identify high-value customers, and explore factors influencing customer engagement. It is suitable for exploratory data analysis (EDA), customer segmentation, and predictive modeling tasks.

From the initial exploration, the dataset contains 255 rows and 14 columns, with minor missing values in the Education and AnnualIncome columns.

```python
df.isnull().sum()
(df.isnull().sum() / len(df)) * 100
```

```python
df['Education'].fillna(df['Education'].mode()[0], inplace=True)
df['AnnualIncome'].fillna(df['AnnualIncome'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

During the preprocessing stage, I examined the dataset for missing values using the .isnull().sum() function. The analysis revealed that the Education and AnnualIncome columns contained 12 missing values each, representing approximately 4.7% of the total dataset.

Since the percentage of missing values was relatively small, I chose to perform imputation rather than dropping rows or columns. For the Education column, which is categorical, I filled missing values using the mode (most frequent category). For the AnnualIncome column, which is numerical, I used the median value to avoid the influence of potential outliers.

Next, I checked for duplicate records using .duplicated(). Any duplicate rows identified were removed to ensure data consistency and reliability.

After preprocessing, the dataset contained no missing values or duplicate records, making it a clean and reliable foundation for further analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
plt.figure()
sns.histplot(df['Age'], bins=10)
plt.title("Distribution of Customer Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

The age distribution shows that most customers fall between approximately 25 and 45 years old. The data appears relatively evenly spread across age groups, with fewer customers at the extreme younger and older ends.

```python
plt.figure()
sns.histplot(df['AnnualIncome'], bins=10)
plt.title("Distribution of Annual Income")
plt.xlabel("Annual Income")
plt.ylabel("Count")
plt.show()
```

The annual income distribution appears fairly spread across different income levels. There are no extreme spikes, suggesting that customers belong to varied income groups rather than a single concentrated bracket.

```python
plt.figure()
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

The gender distribution chart shows the proportion of male and female customers in the dataset. The counts appear relatively balanced, indicating that the dataset does not heavily favor one gender.

```python
plt.figure()
sns.scatterplot(x='AnnualIncome', y='SpendingScore', data=df)
plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

The scatter plot shows how customer spending score varies with annual income. There does not appear to be a strong linear relationship, suggesting that higher income does not necessarily guarantee a higher spending score.

```python
plt.figure()
sns.boxplot(x='Gender', y='LastPurchaseAmount', data=df)
plt.title("Last Purchase Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Last Purchase Amount")
plt.show()
```

The boxplot compares purchase amount distribution across genders. It helps identify differences in median spending and the presence of outliers between male and female customers.

```python
plt.figure()
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

The correlation heatmap shows the strength and direction of relationships between numerical variables. Strong positive correlations indicate that two variables increase together, while negative correlations indicate inverse relationships. Most variables show weak to moderate correlations, suggesting that customer behavior is influenced by multiple independent factors rather than a single dominant feature.
